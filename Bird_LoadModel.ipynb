```python
import numpy as np
import matplotlib.pyplot as plt
import pathlib, os, random
import tensorflow as tf
from keras.models import Sequential
from keras.models import model_from_json
from keras.models import load_model
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator

#Load Model
model = load_model('BirdSpecies_Xception_Jeremy.h5')
```

```python
test_dir = "Bird_Dataset/grading_set/"
```

```python
#Predicting Test Set
def load_and_prep_image(filename, img_shape = 224):
    img = tf.io.read_file(filename) #read image
    #img = filename
    img = tf.image.decode_image(img) # decode the image to a tensor
    img = tf.image.resize(img, size = [img_shape, img_shape]) # resize the image
    img = img/255. # rescale the image
    return img
```

```python
def pred_and_plot(filename, class_names):

    # Import the target image and preprocess it
    img = load_and_prep_image(filename)
    #img = filename

    # Make a prediction
    pred = model.predict(tf.expand_dims(img, axis=0))
    
    # Get the predicted class
    pred_class = class_names[pred.argmax()]

    # Plot the image and predicted class
    plt.imshow(img)
    plt.title(f"Prediction: {pred_class}")
    plt.axis(False)
```

```python
data_dir = pathlib.Path("Bird_Dataset/train")
BirdClasses = np.array(sorted([item.name for item in data_dir.glob("*")]))
#print(BirdClasses)
```

```python
pred_and_plot("Bird_Dataset/grading_set/5.jpg", BirdClasses)
```

```python
pred_and_plot("Bird_Dataset/grading_set/6.jpg", BirdClasses)
```

```python
pred_and_plot("Bird_Dataset/grading_set/8.jpg", BirdClasses)
```
